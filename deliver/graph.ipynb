```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
# Data
df=pd.read_csv("../data/1.csv") 
 

plt.figure(figsize=(9, 5))
# multiple line plot
plt.plot( 'Time', 'NAQM-5', data=df, marker='', color='blue', linewidth=2, label="NAQM-5ms")
plt.plot( 'Time', 'NAQM-20', data=df, marker='', color='green', linewidth=2, label="NAQM-20ms")
plt.plot( 'Time', 'Tail Drop', data=df, marker='', color='red', linewidth=2,label="Tail Drop")
plt.xlabel('Time (s)')
plt.ylabel('Queuing Delay (ms)')
plt.legend()
plt.savefig('1.jpg')
print('Figure 5: Queuing dealy with 30 mbp')
print('---------------------------------------')
print( )
```
